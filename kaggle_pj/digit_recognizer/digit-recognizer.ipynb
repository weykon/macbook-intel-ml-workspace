```python
train_read = pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
train_read.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X_train = train_read.drop(columns=["label"])
Y_train = train_read["label"]
X_train.head()

# label是数字字符，作为目标结果，先分出来
```

```python
Y_train
```

```python
testData = pd.read_csv("/kaggle/input/digit-recognizer/test.csv")
testData.head()
```

```python
X_test = testData
```

```python
X_test
```

```python
KNN = KNeighborsClassifier(n_neighbors=1)
```

```python
KNN.fit(X_train,Y_train)
```

```python
y_predict = KNN.predict(X_test)
```

```python
KNN.score(X_train,Y_train)
```

```python
y_predict
```

```python
ImageId = []
counter = 1
for i in y_predict:
    ImageId.append(counter)
    counter=counter+1
```

```python
finalSubmission = pd.DataFrame({"ImageId":ImageId, "Label":y_predict})
```

```python
finalSubmission
```
